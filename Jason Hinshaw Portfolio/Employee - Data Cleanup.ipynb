```python
import pandas as pd
import numpy as np

employees_raw = pd.read_csv("employees.csv")
employees_raw.head()
```

```python
employees_raw.columns
```

```python
employees_raw.columns = (
    employees_raw.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)
```

```python
employees_raw.columns
```

```python
employees = employees_raw.rename(columns={
    "employee_id": "Employee ID",
    "first_name": "First Name",
    "last_name": "Last Name",
    "role": "Job Title",
    "region": "Region",
    "branch_id": "Branch ID",
    "branch_name": "Branch Location",
    "hire_date": "Hire Date",
    "status": "Employment Status",
    "comp_plan_type": "Compensation Type",
    "base_salary": "Salary",
    "target_bps": "BPS Compensation",
    "manager_id": "Manager ID"
})
```

```python
employees_clean = employees.copy()
employees_clean.head()
```

```python
employees_clean.info()
employees_clean.head()
```

```python
# Convert Hire Date to datetime
employees_clean["Hire Date"] = pd.to_datetime(
    employees_clean["Hire Date"], errors="coerce"
)

# Convert numeric fields
employees_clean["Salary"] = pd.to_numeric(
    employees_clean["Salary"], errors="coerce"
)

employees_clean["BPS Compensation"] = pd.to_numeric(
    employees_clean["BPS Compensation"], errors="coerce"
)

employees_clean["Manager ID"] = pd.to_numeric(
    employees_clean["Manager ID"], errors="coerce"
)
```

```python
employees_clean.loc[
    employees_clean["Compensation Type"] == "salary", "BPS Compensation"
] = 0
```

```python
employees_clean.loc[
    employees_clean["Compensation Type"] == "bps", "Salary"
] = 0
```

```python
employees_clean.head(5)
```

```python
employees_clean["Tenure (Years)"] = (
    (pd.Timestamp("today") - employees_clean["Hire Date"]).dt.days / 365
).round(1)
```

```python
employees_clean["Is BPS Compensation"] = (
    employees_clean["Compensation Type"] == "bps"
).astype(int)
```

```python
employees_clean["Compensation Category"] = employees_clean["Compensation Type"].map({
    "bps": "BPS",
    "salary": "Salary"
})
```

```python
employees_clean["Full Name"] = (
    employees_clean["First Name"] + " " + employees_clean["Last Name"]
)
```

```python
employees_clean["Branch Label"] = (
    employees_clean["Branch Location"] + " - " + employees_clean["Branch ID"].astype(str)
)
```
