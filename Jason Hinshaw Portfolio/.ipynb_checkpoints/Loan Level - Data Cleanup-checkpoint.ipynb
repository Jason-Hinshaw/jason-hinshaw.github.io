```python
loans = loans_raw.rename(columns={
    "loan_number": "Loan Number",
    "application_date": "Application Date",
    "lock_date": "Lock Date",
    "funding_date": "Funding Date",
    "loan_amount": "Loan Amount",
    "property_type": "Property Type",
    "property_state": "Property State",
    "occupancy_type": "Occupancy Type",
    "loan_purpose": "Loan Purpose",
    "channel": "Channel",
    "product_type": "Product Type",
    "rate": "Interest Rate",
    "fico_score": "FICO",
    "cltv": "CLTV",
    "dti": "DTI",
    "lock_term_days": "Lock Term (days)",
    "loan_status": "Loan Status",
    "loan_officer_Id": "Loan Officer ID",
    "account_executive_id": "Account Executive ID",
    "underwriter_id": "Underwriter ID",
    "processor_id": "Processor ID",
    "comp_bps": "Compensation (BPS)",
    "comp_amount": "Compensation ($)",
    "branch_id": "Branch ID",
    "pull_through_flag": "Pull Through Flag",
    "num_extensions": "Extension Count"
})
```

```python
for col in numeric_cols:
    loans_clean[col] = pd.to_numeric(loans_clean[col], errors="coerce")
```

```python
# Define text columns
text_cols = [
    "Property State",
    "Occupancy Type",
    "Loan Purpose",
    "Channel",
    "Product Type",
    "Loan Status"
]

# Clean whitespace
for col in text_cols:
    loans_clean[col] = (
        loans_clean[col]
        .astype(str)
        .str.strip()
        .str.replace(r"\s+", " ", regex=True)
    )

# Standardize casing
for col in text_cols:
    loans_clean[col] = loans_clean[col].str.title()
```

```python
loans_clean.head(10)
```

```python
loans_clean["Is Funded"] = (loans_clean["Loan Status"] == "Funded").astype(int)
```

```python
loans_clean.info()
```

```python
date_cols = ["Application Date", "Lock Date", "Funding Date"]

for col in date_cols:
    loans_clean[col] = pd.to_datetime(loans_clean[col], errors="coerce")
```

```python
loans_clean.info()
```

```python
text_cols = [
    "Property State",
    "Property Type",
    "Occupancy Type",
    "Loan Purpose",
    "Channel",
    "Product Type",
    "Loan Status"
]
```

```python
for col in text_cols:
    loans_clean[col] = (
        loans_clean[col]
        .astype(str)
        .str.strip()
        .str.replace(r"\s+", " ", regex=True)
    )

for col in text_cols:
    loans_clean[col] = loans_clean[col].str.title()
```

```python
loans_clean["Total Time In Process"] = (
    loans_clean["App To Lock Days"].fillna(0) +
    loans_clean["Lock To Fund Days"].fillna(0)
)
```

```python
loans_clean["App To Lock Days"] = (
    loans_clean["Lock Date"] - loans_clean["Application Date"]
).dt.days
```

```python
loans_clean["Lock To Fund Days"] = (
    loans_clean["Funding Date"] - loans_clean["Lock Date"]
).dt.days
```

```python
loans_clean["Days To Fund"] = (
    loans_clean["Funding Date"] - loans_clean["Application Date"]
).dt.days
```

```python
bins = [0, 579, 619, 659, 699, 739, 799, 850]
labels = [
    "Poor (<580)",
    "Fair (580-619)",
    "Near Prime (620-659)",
    "Prime (660-699)",
    "Prime+ (700-739)",
    "Excellent (740-799)",
    "Elite (800+)"
]

loans_clean["FICO Bucket"] = pd.cut(loans_clean["FICO"], bins=bins, labels=labels)
```

```python
bins = [0, 60, 70, 80, 90, 100, 200]
labels = [
    "<=60%",
    "61-70%",
    "71-80%",
    "81-90%",
    "91-100%",
    "100%+"
]

loans_clean["CLTV Bucket"] = pd.cut(loans_clean["CLTV"], bins=bins, labels=labels)
```

```python
loans_clean["High DTI Flag"] = (loans_clean["DTI"] > 45).astype(int)
```
